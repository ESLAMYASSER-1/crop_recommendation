# DNN

```python
import torch 
import torchvision.transforms as trns
from torch.utils.data import Dataset, DataLoader
from tqdm import tqdm
device= torch.device("cuda" if torch.cuda.is_available() else "cpu")
```

```python
data = pd.read_csv("crop_recommendation.csv")
data["label"] = le.transform(data["label"])
data.head()
```

```python
class CSV_DATA(Dataset):
    def __init__(self, csv_file, transform=None):
        
        self.data_frame = csv_file
        self.transform = transform

    def __len__(self):
        return len(self.data_frame)

    def __getitem__(self, idx):
        row = self.data_frame.iloc[idx]
        data = torch.tensor(row[:-1].values.astype(float))  
        label = torch.tensor(row[-1])  
        
        if self.transform:
            data = self.transform(data)
        
        return data, label
```

```python
csv_data = CSV_DATA(data)
dataloader = DataLoader(csv_data, batch_size=64, shuffle=True)
```

```python
class NeuralNet(torch.nn.Module):
    def __init__(self, device):
        super(NeuralNet, self).__init__()
        self.fc1 = torch.nn.Linear(7, 64).to(device)  
        self.fc2 = torch.nn.Linear(64, 30).to(device)  
        self.fc3 = torch.nn.Linear(30, 22).to(device)  
        
    def forward(self, x):
        x = torch.relu(self.fc1(x))  
        x = torch.relu(self.fc2(x))  
        x = self.fc3(x)
        return x
```

```python
DNN = NeuralNet(device)
criterion = torch.nn.CrossEntropyLoss()
optimizer = torch.optim.Adam(DNN.parameters(), lr= 0.01)
```

```python
from tqdm import tqdm
import torch

def train(model, dataloader, criterion, optimizer, n_epochs, device, print_=True):
    model.to(device)
    loss_l = []
    accuracy_l = []  
    
    print("First epoch could take several minutes...")
    for epoch in tqdm(range(n_epochs)):
        model.train()  
        epoch_loss = 0.0
        for X, y in dataloader:
            X, y = X.to(device), y.to(device)
            optimizer.zero_grad()
            X = X.float()
            y = y.long()  
            predict = model(X)
            
            loss = criterion(predict, y)
            loss.backward()
            optimizer.step()
            epoch_loss += loss.item()  
            
        epoch_loss /= len(dataloader)  
        loss_l.append(epoch_loss)
        
        if print_:
            print(f"Epoch: {epoch+1} Done! | Average Epoch Loss: {epoch_loss:.4f}")
            print('Learning Rate:', optimizer.param_groups[0]['lr'])
    
    return loss_l, model
```

```python
Loss, model = train(DNN, dataloader, criterion, optimizer, 20, device)
```

```python
DNN.eval()
```
